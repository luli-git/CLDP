```python
import pandas as pd
```

```python
diagnosis = pd.read_csv(
    "diagnoses_icd.csv/diagnoses_icd.csv"
)
```

```python
prescriptions = pd.read_csv(
    "prescriptions.csv/prescriptions.csv"
)
```

```python
prescriptions
```

```python
# filter only icd_code equals 3310 or G30, G300, G301, G308, G309

diagnosis_ad = diagnosis[
    diagnosis.icd_code.isin(
        [
            "3310",
            "G30",
            "G300",
            "G301",
            "G308",
            "G309",
        ]
    )
]
```

```python
px_ad_related = prescriptions[
    prescriptions.hadm_id.isin(diagnosis_ad.hadm_id)
]
print(px_ad_related.shape)
px_ad_related[
    px_ad_related.drug_type != "MAIN"
].shape
```

```python
# filter diagnosis whose hadm_id is in diagnosis_ad hadm_id

diag_ad_related = diagnosis[
    diagnosis.hadm_id.isin(
        diagnosis_ad.hadm_id
    )
]
```

```python
# get the number of unique patients
print(
    "Number of unique patients: {}".format(
        diagnosis_ad.subject_id.nunique()
    )
)
```

```python
# on average, how many diagnoses per patient?
print(
    "Average number of diagnoses per patient: {}".format(
        diagnosis_ad.groupby("subject_id").size().mean()
    )
)
```
